```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

```python
model_name = "deepset/roberta-base-squad2"
```

```python
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
nlp = pipeline("question-answering",model=model,tokenizer=tokenizer)
```

```python
corpus = ["It was a beautiful morning for a walk",
          "A gentle breeze rustled the leaves of the tall trees.",
         "Sunlight filtered through the dense canopy, dappling the forest floor in shifting patterns.",
          "A small squirrel darted across the path, its bushy tail held high.",
          "The air was crisp and smelled of pine and damp earth.",
          "Nearby, a bird sang a cheerful melody.",
          "The path wound along a small creek, its water gurgling softly over smooth stones.",
          "A tiny fish swam in the clear stream.",
          "A dragonfly, its wings iridescent, hovered near the water's edge.",
          "The feeling of peace was all-consuming.",
          "Continuing on, the trail opened into a wide meadow.",
          "Wildflowers of every color swayed in the breeze.",
          "A monarch butterfly landed on a bright purple thistle.",
          "The distant mountains were a deep shade of blue.",
          "Wispy clouds drifted lazily across the sky.",
          "Sitting on a rock, the world was observed for a while.",
          "The experience felt like a small reset.",
          "A sense of deep calm settled.",
          "Feeling refreshed and grateful for the quiet moments in nature.",
          "It was the perfect start to the day."]
```

```python
while True:
  user_input = input("Ask a question, press q to exit:")
  if user_input.lower() =='q':
    break
  input_to_model = {"question":user_input,"context":" ".join(corpus)}
  print("Answer: ",nlp(input_to_model)['answer'])
```
